```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
import os 
%matplotlib inline
```

```python
path=r'/dodrio/scratch/projects/2022_200/project_output/bclimate/sdeherto/postprocessing/'

#2015 irr fluxes
case='b.e213.BSSP126_BPRP.f09_g17.esm-futctl-i308'
ens='001'
fn_ctl1= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futineq-i308'
fn_ineq1= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futsust-i308'
fn_sust1= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'

ens='002'
fn_ctl2= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futineq-i308'
fn_ineq2= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futsust-i308'
fn_sust2= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
ens='003'
fn_ctl3= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futineq-i308'
fn_ineq3= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
case='b.e213.BSSP126_BPRP.f09_g17.esm-futsust-i308'
fn_sust3= case+'.'+ens+'/QIRRIG_'+case+'.'+ens+'.clm2.h0.nc'
```

```python
#read in arrays

ds_ctl1=xr.open_dataset(path+fn_ctl1)
da_ctl1_2015=ds_ctl1['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ctl1_2100=ds_ctl1['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_ctl2=xr.open_dataset(path+fn_ctl2)
da_ctl2_2015=ds_ctl2['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ctl2_2100=ds_ctl2['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_ctl3=xr.open_dataset(path+fn_ctl3)
da_ctl3_2015=ds_ctl3['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ctl3_2100=ds_ctl3['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

da_ctl_2015=da_ctl1_2015+da_ctl2_2015+da_ctl3_2015
da_ctl_2100=da_ctl1_2100+da_ctl2_2100+da_ctl3_2100

ds_ineq1=xr.open_dataset(path+fn_ineq1)
da_ineq1_2015=ds_ineq1['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ineq1_2100=ds_ineq1['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_ineq2=xr.open_dataset(path+fn_ineq2)
da_ineq2_2015=ds_ineq2['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ineq2_2100=ds_ineq2['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_ineq3=xr.open_dataset(path+fn_ineq3)
da_ineq3_2015=ds_ineq3['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_ineq3_2100=ds_ineq3['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

da_ineq_2015=da_ineq1_2015+da_ineq2_2015+da_ineq3_2015
da_ineq_2100=da_ineq1_2100+da_ineq2_2100+da_ineq3_2100

ds_sust1=xr.open_dataset(path+fn_sust1)
da_sust1_2015=ds_sust1['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_sust1_2100=ds_sust1['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_sust2=xr.open_dataset(path+fn_sust2)
da_sust2_2015=ds_sust2['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_sust2_2100=ds_sust2['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

ds_sust3=xr.open_dataset(path+fn_sust3)
da_sust3_2015=ds_sust3['QIRRIG'][0:12*30,:,:].mean('time')*31536000
da_sust3_2100=ds_sust3['QIRRIG'][-12*30:,:,:].mean('time')*31536000   ##mm/year

da_sust_2015=da_sust1_2015+da_sust2_2015+da_sust3_2015
da_sust_2100=da_sust1_2100+da_sust2_2100+da_sust3_2100
```

```python
da_ctl_2015.plot(vmin=0,vmax=300,cmap='GnBu')
```

```python
da_sust_2100.plot(vmin=0,vmax=300,cmap='GnBu')
```

```python
da_ineq_2100.plot(vmin=0,vmax=300,cmap='GnBu')
```

```python
diff_scen=da_sust_2100-da_ineq_2100
diff_scen.plot(vmin=-100,vmax=100,cmap='coolwarm')
```

```python
diff_ineq=da_ineq_2100-da_ctl_2100
diff_ineq.plot(vmin=-100,vmax=100,cmap='coolwarm')
```

```python
diff_sust=da_sust_2100-da_ctl_2100
diff_sust.plot(vmin=-100,vmax=100,cmap='coolwarm')
```

```python
projection = ccrs.PlateCarree()

# Create a figure with 3 subplots in a horizontal layout
fig, axes = plt.subplots( 3,1, figsize=(8, 20), subplot_kw={'projection': projection})

da_list=[da_ctl_2015,diff_sust,diff_ineq]
title_list=['CTL 2015', 'sustainability', 'inequality']
label_list=['Extent (mm/day)', 'Change (mm/day)','Change (mm/day)']
# Plot the data with pcolormesh for each subplot
for i, ax in enumerate(axes):
    if i==0:
        pcm = ax.pcolormesh(ds_ctl2.lon, ds_ctl2.lat, da_list[i], vmin=0,vmax=250,cmap='GnBu',transform=projection)
    else:
        pcm = ax.pcolormesh(ds_ctl2.lon, ds_ctl2.lat, da_list[i], vmin=-100,vmax=100,cmap='bwr_r',transform=projection)
    # Add coastlines for reference
    ax.coastlines()

    # Add a colorbar
    cbar = plt.colorbar(pcm, ax=ax, orientation='vertical',shrink=0.9)
    cbar.set_label(label_list[i])

    # Set axis labels
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title(title_list[i])
    ax.set_xlim(-180, 180)
# Adjust layout to prevent overlap
fig.subplots_adjust(left=0.01, right=0.99,bottom=0., top=0.51,hspace=0., wspace=0.)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
import cartopy.crs as ccrs

# Define the colors for the custom colormap
colors = [(1, 1, 1), (0, 0, 1)]  # White to Blue

# Create the colormap
custom_cmap = LinearSegmentedColormap.from_list('custom_white_blue', colors)

# Create a figure with 3 subplots in a horizontal layout
fig, axes = plt.subplots(3, 1, figsize=(8, 20), subplot_kw={'projection': ccrs.PlateCarree()})

da_list = [da_ctl_2015, diff_sust, diff_ineq]
title_list = ['CTL 2015', 'sustainability', 'inequality']
label_list = ['Extent (mm/day)', 'Change (mm/day)', 'Change (mm/day)']

# Plot the data with pcolormesh for each subplot
for i, ax in enumerate(axes):
    if i == 0:
        pcm = ax.pcolormesh(ds_ctl2.lon, ds_ctl2.lat, da_list[i], vmin=0, vmax=250, cmap=custom_cmap, transform=ccrs.PlateCarree())
    else:
        pcm = ax.pcolormesh(ds_ctl2.lon, ds_ctl2.lat, da_list[i], vmin=-100, vmax=100, cmap='bwr_r', transform=ccrs.PlateCarree())
    
    # Add coastlines for reference
    ax.coastlines()

    # Add a colorbar
    cbar = plt.colorbar(pcm, ax=ax, orientation='vertical', shrink=0.9)
    cbar.set_label(label_list[i])

    # Set axis labels
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title(title_list[i])

    
# Adjust layout to prevent overlap
fig.subplots_adjust(left=0.01, right=0.99, bottom=0., top=0.51, hspace=0., wspace=0.)

# Show the plot
plt.show()
```
